Furthermore, calling df.dropna() to eliminate missing values because the values in the column are strings, this notebook prefers to just remove it altogether. 

```python
df.info()
```

Then call df.info() to check whether the missing value is gone. This is evidenced by the value contained in the Non-Null Count which all have the same value, which means that there are no more missing values. 

```python
#Drop Column
df.drop(labels=['Type 2', 'Legendary'], axis=1, inplace=True)
df
```

To remove unused columns, then call df.drop to remove Type 2 and Legendary columns because this notebook will not discuss the two columns.
